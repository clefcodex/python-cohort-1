 ðŸ”¹ 3. Wallet Analysis
# Compute total amount sent and received by each wallet.

# Identify the top 5 wallets by total amount sent (group + sum + sort).

# Create a new column: net_flow = received - sent.

```python
# Compute total amount sent and received by each wallet.

# Total sent per wallet
sent = df.groupby('sender_wallet')['amount'].sum().reset_index()
sent.rename(columns={'amount': 'total_sent'}, inplace=True)

# Total received per wallet
received = df.groupby('receiver_wallet')['amount'].sum().reset_index()
received.rename(columns={'amount': 'total_received'}, inplace=True)

# Merge sent and received into one table
wallets = pd.merge(sent, received, left_on='sender_wallet', right_on='receiver_wallet', how='outer')

# Fill NaNs with 0 (for wallets that only sent or only received)
wallets['sender_wallet'] = wallets['sender_wallet'].combine_first(wallets['receiver_wallet'])
wallets.drop(columns='receiver_wallet', inplace=True)
wallets.rename(columns={'sender_wallet': 'wallet'}, inplace=True)
wallets.fillna(0, inplace=True)

# Optional: convert columns to consistent type
wallets['total_sent'] = wallets['total_sent'].astype(float)
wallets['total_received'] = wallets['total_received'].astype(float)

# Show result
print(wallets)
```

```python
# Identify the top 5 wallets by total amount sent (group + sum + sort).

top_senders = wallets.sort_values(by=['total_sent'], ascending=False)

top_senders.head(5)
```

```python
# Create a new column: net_flow = received - sent.

wallets['net_flow'] = wallets['total_received'] - wallets['total_sent']

wallets.sort_values(by=['net_flow'], ascending=False)
```

4. Token Analysis
Find the average, median, and standard deviation of amount per token (use .agg()).

Identify the token with the highest transaction volume (in total amount, not count).

Plot a bar chart showing total amount transferred per token using matplotlib.

```python
# Find the average, median, and standard deviation of amount per token (use .agg()).

stats = df.groupby('token')['amount'].agg(
    average='mean',
    median='median',
    std_dev='std'
).reset_index()

print(stats)
```

```python
# Identify the token with the highest transaction volume (in total amount, not count).

# Group by token and sum the amount
volume_per_token = df.groupby('token')['amount'].sum().reset_index()

# Sort in descending order to find the highest
top_token = volume_per_token.sort_values(by='amount', ascending=False).head(1)

print(top_token)
```

ðŸ”¹ 5. Time-based Analysis
Group data by day (timestamp.dt.date) and compute:

Number of transactions per day

Total volume per day

Average transaction size per day

```python
# Ensure 'timestamp' column is in datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract date from timestamp
df['date'] = df['timestamp'].dt.date

# Group by date
daily_stats = df.groupby('date').agg(
    transactions_per_day=('amount', 'count'),
    total_volume_per_day=('amount', 'sum'),
    average_tx_size_per_day=('amount', 'mean')
).reset_index()

print(daily_stats.head())
```

ðŸ”¹ 7. Advanced Challenges
1. Create a pivot table: Rows = Token, Columns = Day, Values = Total amount transferred

2. Use apply() to tag transactions as:

    low if amount < 100

    medium if 100 <= amount <= 1000

    high if amount > 1000

3. Use np.where() to create a binary feature: is_large_tx (1 if amount > 1000, else 0)

```python
# 1. Create a pivot table: Rows = Token, Columns = Day, Values = Total amount transferred

df['date'] = pd.to_datetime(df['timestamp']).dt.date  # extract just the date

pivot_table = df.pivot_table(
    values='amount',
    index='token',
    columns='date',
    aggfunc='sum',
    fill_value=0
)

print(pivot_table)
```

```python
# 2. Use apply() to tag transactions as:

#     low if amount < 100

#     medium if 100 <= amount <= 1000

#     high if amount > 1000


def tag_amount_level(amount):
    if amount < 100:
        return 'low'
    elif amount <= 1000:
        return 'medium'
    else:
        return 'high'

df['amount_level'] = df['amount'].apply(tag_amount_level)
print(df[['amount', 'amount_level']].head())
```

```python
# 3. Use np.where() to create a binary feature: is_large_tx (1 if amount > 1000, else 0)


df['is_large_tx'] = np.where(df['amount'] > 1000, 1, 0)
print(df[['amount', 'is_large_tx']].head())
```
